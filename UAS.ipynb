```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Load the image

```python
  image = cv2.imread(image_path)
```

# Convert the image to grayscale

```python
 gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

   # Apply a Gaussian blur to reduce noise

```python
blur = cv2.GaussianBlur(gray, (5, 5), 0)
```

# Perform edge detection using the Canny algorithm


```python
 edges = cv2.Canny(blur, 40, 150)
```

 # Find contours in the image

```python
 contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

# Create a mask for the contours

```python
mask = np.zeros_like(gray)
cv2.drawContours(mask, contours, -1LIL, (200), thickness=cv2.FILLED)
```

# Apply the mask to the original image

```python
result = cv2.bitwise_and(image, image, mask=mask)
```

# Display the original image and the result

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')
axs[0].axis('off')
axs[1].imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
axs[1].set_title('Image with Background Removed')
axs[1].axis('off')
plt.show()
```
